```python
import pandas as pd

# Load the two Excel files
real_file = "results/realheartfinal.xlsx"
fake_file = "lipfd_real.xlsx"

real_df = pd.read_excel(real_file)
fake_df = pd.read_excel(fake_file)

# Assume the common column for matching is named 'Identifier'
# Ensure the column names align with your actual files
key_column = 'File Name'
real_score_column = 'Score'  # Adjust if your real score column has a different name
fake_score_column = 'Original Score'  # Adjust if your fake score column has a different name

# Adjust the real scores
real_df[real_score_column] = 1 - real_df[real_score_column]

# Merge the dataframes on the identifier
merged_df = pd.merge(real_df[[key_column, real_score_column]],
                      fake_df[[key_column, fake_score_column]],
                      on=key_column,
                      suffixes=('_Real', '_Fake'))

# Save to a new Excel file
output_file = "real_merged_scores.xlsx"
merged_df.to_excel(output_file, index=False)

print(f"Merged file saved as {output_file}")
```
